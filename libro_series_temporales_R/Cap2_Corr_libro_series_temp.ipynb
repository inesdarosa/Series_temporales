```r
data(AirPassengers)
AP <- AirPassengers

# Aquí la serie esta descompuesta usando la función decompose(), en tendencia, estacinalidad y parte randómica

AP.decom <- decompose(AP, "multiplicative")
```

```r
plot(AP.decom)
```

```r
plot(ts(AP.decom$random[7:138]))# ploteo del componente randómico separado del resto.

acf(AP.decom$random[7:138])# ploteo del correlograma del componente randómico

# The correlogram in Figure suggests a damped cosine shape that is characteristic of an autoregressive model of order 2
# El correlograma muestra un resultado consistente con un modelo autoregresivo de orden 2.
```

```r
sd(AP[7:138])# desvío en la serie original

sd(AP[7:138] - AP.decom$trend[7:138])# desvío luego de remover la tendencia

sd(AP.decom$random[7:138])# desvío luego del ajuste estacional, muestra una gran reducción 
                          # lo que sugiere que el ajuste estacional fue efectivo.
```
